# Image to Arrow Table Conversion



```python
# preload needed libraries 

import pyarrow as pa
import pandas as pd

# import sys
 
# # setting path - only needed when running this from the `with-code` folder.
# sys.path.append('../')

from CVDemoUtils import CVDemo
cvDemo = CVDemo()
cvDemo.COCO_CLASSES_PATH = "./models/coco_classes.pickle"

# used to retrieve files from a directory
import os
```

```python
# set variables here

width = 640
height = 480

image_path = './images/input/example'
df_path = './df/input/example'
arrow_path = './arrow/input/example'
```

```python
# retrieve the DataFrame version of each file and export to Arrow

files = os.listdir(image_path)
files = [f for f in files if os.path.isfile(image_path+'/'+f)] #Filtering only the files.
display(files)

for file in files:
    # if file.endswith(".png"):
    display(file)
    file_path = f'{image_path}/{file}'
    df_file_path = f'{df_path}/{file}.df.json'
    arrow_file_path = f'{arrow_path}/{file}.arrow'
    
    dfImage, resizedImage = cvDemo.loadImageAndConvertToDataframe(file_path, 
                                                            width, 
                                                            height
                                                            )
    display(dfImage)
    # export dfImage to a file
    os.makedirs(os.path.dirname(df_file_path), exist_ok=True)
    dfImage.to_json(df_file_path, orient="records")
    # read from that file again
    df = pd.read_json(df_file_path, orient="records")
    # save the arrow table
    test_table = pa.Table.from_pandas(df)
    display(test_table)
    os.makedirs(os.path.dirname(arrow_file_path), exist_ok=True)
    # # save to a file
    with pa.OSFile(arrow_file_path, 'wb') as sink:
        with pa.ipc.new_file(sink, test_table.schema) as arrow_ipc:
            arrow_ipc.write(test_table)
            arrow_ipc.close()


# dfImage, resizedImage = cvDemo.loadImageAndConvertToDataframe(image, 
#                                                               width, 
#                                                               height
#                                                               )
```
